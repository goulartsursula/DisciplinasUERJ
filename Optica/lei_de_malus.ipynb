```
import numpy as np
import math
import matplotlib.pyplot as plt

def converterGrausParaRad(numero):
  rad = (numero/180)*math.pi
  return rad

def coseno(rad):
  numero = converterGrausParaRad(rad)
  cont = 0
  resultado = 1 # ERRO!!!
  while(cont < 50):
    cont += 1
    resultado += (((-1)**cont)*(numero**(2 * cont)))/(math.factorial(2 * cont))

    return round(resultado, 6) # MELHORIA

angles = [i for i in range(0, 181, 10)]
for angle in angles:
  cos = coseno(angle)
  cos_ = math.cos(converterGrausParaRad(angle))
  cos_quadrado = np.power(cos_, 2)
  intensity = 0.5*cos_quadrado
  print('ângulo: {} cosseno: {:.3f} ({:.3f}) cos²: {:.3f} intensidade: {:.3f})' \
                .format(angle, cos, cos_, cos_quadrado, intensity))

# gráfico das intensidades contra os ângulos 
angulo = np.arange(0, 181, 10)
intensidade = ([0.500, 0.485, 0.442, 0.375, 0.293, 0.207, 0.125, 0.058, 0.015, 0.000, 0.015, 0.058, 0.125, 0.207, 0.293, 0.375, 0.442, 0.485, 0.500])
plt.xlabel(r'Ângulo de polarização $(\theta)$', fontsize = 13)
plt.ylabel(r'Intensidade luminosa $\frac{W}{m^2}$', fontsize = 13)
plt.title(r'Intensidade luminosa $\times$ Ângulo de polarização' '\n', fontsize = 15)
plt.scatter(angulo, intensidade, color = 'r')
plt.plot(angulo, intensidade, color = 'r')
plt.grid(True)
plt.show()

#  Gráfico das intensidades contra cos²(theta)
coseno_quadrado = ([1.000, 0.970,  0.883, 0.750, 0.587, 0.413, 0.250, 0.117, 0.030, 0.000, 0.030, 0.117, 0.250, 0.413, 0.587, 0.750, 0.883, 0.970, 1.000])
intensidade = ([0.500, 0.485, 0.442, 0.375, 0.293, 0.207, 0.125, 0.058, 0.015, 0.000, 0.015, 0.058, 0.125, 0.207, 0.293, 0.375, 0.442, 0.485, 0.500])
plt.xlabel(r'$\cos²(\theta)$', fontsize = 13)
plt.ylabel(r'Intensidade luminosa $\frac{W}{m^2}$', fontsize = 13)
plt.title(r'Intensidade luminosa $\times \cos²(\theta)$' '\n', fontsize = 15)
plt.scatter(coseno_quadrado, intensidade, color = 'r')
plt.plot(coseno_quadrado, intensidade, color = 'r')
plt.grid(True)
plt.show()

#gráfico log-log
coseno_quadrado = ([1.000, 0.970,  0.883, 0.750, 0.587, 0.413, 0.250, 0.117, 0.030, 0.000, 0.030, 0.117, 0.250, 0.413, 0.587, 0.750, 0.883, 0.970, 1.000])
intensidade = ([0.500, 0.485, 0.442, 0.375, 0.293, 0.207, 0.125, 0.058, 0.015, 0.000, 0.015, 0.058, 0.125, 0.207, 0.293, 0.375, 0.442, 0.485, 0.500])
plt.xlabel(r'log($\cos²(\theta)$)', fontsize = 13)
plt.ylabel('log(I)', fontsize = 13)
plt.title(r'log(I) $\times log(\cos²(\theta)$)' '\n', fontsize = 15)
plt.loglog(coseno_quadrado, intensidade, color = 'r')
plt.scatter(coseno_quadrado, intensidade, color = 'r')
plt.grid(True)
plt.show()
```
